```python
#mengubah tipe data menjadi datetime untuk beberapa kolom di orders
datetime_orders = ['order_purchase_timestamp','order_approved_at','order_delivered_carrier_date','order_delivered_customer_date','order_estimated_delivery_date']
for column in datetime_orders:
    data['orders'][column] = pd.to_datetime(data['orders'][column])

#mengubah shipping_limit_date menjadi tipe date datetime pada items
data['items']['shipping_limit_date'] = pd.to_datetime(data['items']['shipping_limit_date'])

#mengubah tipe data menjadi datetime untuk beberapa kolom di reviews
datetime_reviews = ['review_creation_date','review_answer_timestamp']
for column in datetime_reviews:
    data['reviews'][column] = pd.to_datetime(data['reviews'][column])
```

### Cleaning Data

- **Orders data**

```python
# mencetak jumlah status pesanan dalam data
print(data['orders']['order_status'].value_counts())
```

```python
print('Pesanan yang telah dikirimkan: ' + str(round((data['orders'].order_status.value_counts()['delivered']/len(data['orders']))*100,2)) + "% dari data.")
```

```python
#cari semua data order_id yang order statusnya delivered kemudian filter
#semua dataframe yang memiliki order_id sebagai foreign key

delivered = data['orders']['order_id'][data['orders']['order_status'] != 'delivered'].values

for df_name, df in data.items():
    if 'order_id' in df.columns:
        df.drop(df.loc[df['order_id'].isin(delivered)].index, inplace = True)
```

```python
#cek nilai NaN pada setiap dataset
for df_name, df in data.items():
    for col in df.columns:
        if df[col].isnull().sum():
            print(col + '(%) : '+str((sum(df[col].isnull())/len(df))*100))
```

```python
# Memeriksa nilai NaN dalam kolom tertentu dalam dataframe pesanan
missing_delivery_dates = data['orders'].order_delivered_customer_date.isnull().sum()
print( str(missing_delivery_dates) +
      ' pesanan tidak ada tanggal pengiriman')
```

```python
# Kita dapat menghapusnya dengan cara yang sama seperti sebelumnya
# pertama dapatkan semua order_id di mana status pesanannya telah dikirim
null_deliveries = data['orders']['order_id'][data['orders']['order_delivered_customer_date'].isnull()].values

# sekarang, filter setiap dataframe lain di mana order_id adalah kunci asing
for df_name, df in data.items():
    if 'order_id' in df.columns:
        df.drop(df.loc[df['order_id'].isin(null_deliveries)].index, inplace=True)
```

```python
# Jika pesanan dikirimkan sebelum atau pada tanggal perkiraan pengiriman, maka anggap 'On Time', jika tidak, anggap 'Late'.
data['orders']['delivered_on_time'] = np.where(data['orders']['order_delivered_customer_date'] <= data['orders']['order_estimated_delivery_date'], 'On Time', 'Late')
```

## Visualization & Explanatory Analysis

### Pertanyaan 1: Produk mana yang paling laku?

```python
# Menggabungkan dataset 'items' dengan dataset 'products' berdasarkan kolom 'product_id' dengan metode inner join
items_product = data['items'].merge(data['products'], on='product_id', how='inner')

# Menggabungkan hasilnya dengan dataset 'orders' berdasarkan kolom 'order_id' dengan metode inner join
orders_ip = data['orders'].merge(items_product, on='order_id', how='inner')

# Membuat tabel pivot yang mengagregasi berdasarkan jumlah item yang dibeli dan rata-rata harga produk
product_revenue = orders_ip.pivot_table(index=['product_id'], aggfunc={'order_item_id': 'sum', 'price': 'mean'})
product_revenue['total'] = product_revenue['order_item_id'] * product_revenue['price']
product_revenue.rename(columns={'order_item_id': 'sell_probability'}, inplace=True)
product_revenue['sell_probability'] = product_revenue['sell_probability'] / len(product_revenue)
product_revenue.sort_values(by='total', ascending=False)
```

* Membuat plotting probabilitas penjualan produk, harga, dan pendapatan.

```python
# Fungsi kustom untuk menetapkan desain plot
def custom_plot(ax, spines):
    for loc, spine in ax.spines.items():
        if loc in spines:
            spine.set_position(('outward', 10))
        else:
            spine.set_color('none')
    ax.xaxis.set_ticks_position('bottom')
    ax.yaxis.set_ticks_position('left')

x = np.log(product_revenue.sell_probability)
y = np.log(product_revenue.price)

fig, ax = plt.subplots(figsize=(8, 6))
custom_plot(ax, ['bottom', 'left'])

plt.title('Harga Produk vs. Probabilitas Penjualan', fontsize=16)
plt.xlabel('Log Probabilitas Penjualan', fontsize=12)
plt.ylabel('Log Harga Produk', fontsize=12)

plt.xlim(-11, -3)
plt.ylim(0, 9)

plt.yticks(range(10), [int(np.exp(x)) for x in range(10)], fontsize=10)
plt.xticks(range(-10, -2), [round(np.exp(x), 4) for x in range(-10, -2)], fontsize=10, rotation=30)

# Menggunakan colormaps yang berbeda dan menetapkan jumlah grid
hb = ax.hexbin(x, y, gridsize=14, C=product_revenue.total, reduce_C_function=np.sum, cmap='viridis') # Ubah ke colormap 'viridis'

# Menambahkan colorbar
cb = fig.colorbar(hb, ax=ax)
cb.set_label('Pendapatan Produk (R$)', rotation=270, labelpad=20, fontsize=12)

# Menyesuaikan tata letak dan menyimpan gambar
plt.tight_layout()
plt.savefig('data/product_probability_sum_custom.png', dpi=300)

plt.show()
```


*Kita dapat mengatakan bahwa produk-produk yang menghasilkan pendapatan terbanyak adalah barang-barang dengan harga menengah yang sering terjual.*

### Pertanyaan 2: Berapa rata rata uang yang dihabiskan oleh kustomer?

```python
# merge dataset pelanggan dengan dataset orders dan payments
pay_ord_cust = orders.merge(payments, on='order_id', how='outer').merge(customers, on='customer_id', how='outer')

# Menghitung customer spent
customer_spent = pay_ord_cust.groupby('customer_unique_id').agg({'payment_value': 'sum'}).sort_values(by='payment_value', ascending=False)

# Menghitung rata-rata cutomer spent dan standar deviasi
customer_mean = customer_spent['payment_value'].mean()
customer_std = stats.sem(customer_spent['payment_value'])

# Memprediksi rata-rata populasi dari sampel
stats.t.interval(0.95, loc=customer_mean, scale= customer_std, df= len(customer_spent)- 1)
```

```python
# Menghitung rata-rata dan interval kepercayaan untuk pelanggan di setiap wilayah
customer_regions = pay_ord_cust.groupby('customer_state').agg({'payment_value':['mean', 'std'], 'customer_unique_id':'count'})
customer_regions.reset_index(inplace=True)

cis = stats.t.interval(0.95, loc=customer_regions['payment_value']['mean'], scale=customer_regions['payment_value']['std']/np.sqrt(customer_regions['customer_unique_id']['count']), df=customer_regions['customer_unique_id']['count'] - 1)

customer_regions['ci_low'] = cis[0]
customer_regions['ci_hi'] = cis[1]
```

```python

# matplotlib adjustments
def default_plot(ax, spines):
    ax = plt.gca()
    # Hapus sumbu yang tidak perlu
    ax.spines["top"].set_visible(False)
    ax.spines["right"].set_visible(False)

    ax.get_yaxis().set_tick_params(direction='out', width=1.5) 
    ax.get_xaxis().set_tick_params(direction='out', width=1.5) 

    for loc, spine in ax.spines.items():
        if loc in spines:
            spine.set_position(('outward', 10)) 

    if 'left' in spines:
        ax.yaxis.set_ticks_position('left')

    if 'right' in spines:
        ax.yaxis.set_ticks_position('right')

    if 'bottom' in spines:
        ax.xaxis.set_ticks_position('bottom')

    return ax

# Urutkan secara ascending agar plot yang lebih mudah dibaca
plot = customer_regions.sort_values(by=('payment_value', 'mean'))

# buat figurenya
fig, ax = plt.subplots(figsize=(12, 4))
ax = default_plot(ax, ['left', 'bottom'])
plt.xticks(rotation=30)
plt.xlabel('State')
plt.ylabel('Rata-rata Transaksi (95% CI)')
plt.xlim(-0.5, 26.5)
plt.ylim(125, 325)
plt.scatter(plot['customer_state'], plot['payment_value']['mean'], s=100, c=plot['payment_value']['mean'], cmap='viridis') # Ubah pola warnanya
plt.vlines(plot['customer_state'], plot['ci_low'], plot['ci_hi'], lw=0.5)
plt.tight_layout()
plt.savefig('data/customers_ci.png', dpi=300)
```
